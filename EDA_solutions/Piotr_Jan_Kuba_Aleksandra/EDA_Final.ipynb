## Wniosek
* Status jest pomijalnie zależny od Credit_score i Gender

```python
BCstatus = data.groupby('business_or_commercial')['Status'].value_counts()

bcProc = BCstatus.agg('b/c')[0] / (BCstatus.agg('b/c')[0] + BCstatus.agg('b/c')[1] )*100

nobcProc = BCstatus.agg('nob/c')[0] / (BCstatus.agg('nob/c')[0] + BCstatus.agg('nob/c')[1] )*100

print(f'Procent spłaceń kredytu w celu business/commercial: {bcProc}%')
print(f'Procent spłaceń kredytu dla w celu prywatnym: {nobcProc}%')
```

## Wniosek
* Kredyty wzięte w celach prywatnych są częściej spłacane niż wzięte w celach biznesowych

```python
sns.boxplot(x = data['business_or_commercial'],y = data['loan_amount'], showfliers = False,hue=data['Status'])
```

## Winoski
* Częściej spłacane są wyższe kredyty (możliwe, że ludzie zapominają o niższych)
* Osoby prywatne biorą większe kredyty

```python
data.groupby('Credit_Worthiness')['Status'].value_counts()
CWstatus = data.groupby('Credit_Worthiness')['Status'].value_counts()


l1Proc = CWstatus.agg('l1')[0] / (CWstatus.agg('l1')[0] + CWstatus.agg('l1')[1] )*100

l2Proc = CWstatus.agg('l2')[0] / (CWstatus.agg('l2')[0] + CWstatus.agg('l2')[1] )*100

print(f'Procent spłaceń kredytu dla Credit_Worthiness: {l1Proc}%')
print(f'Procent spłaceń kredytu dla nie Credit_Worthiness: {l2Proc}%')
```

## Wnioski
* Dla l1 są częsciej spłacane kredyty niż dla l2

# Wnioski
* Joint (łączone kredyty) częściej spłacają od reszty

```python
# Wybór zmiennych ilościowych (tylko liczby)
numeric_columns = data.select_dtypes(include=['float64', 'int64']).drop(columns=['ID', 'year', 'Status'])

# Obliczenie macierzy korelacji
correlation_matrix = numeric_columns.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

## Wnioski
* Silną korelacje widać między loan_amount a property_value, równiez między interest_rate_spread a rate_of_interest
* Ewentualnie mozna jeszcze rozwazyć wyrózniającą korelację między income a loan_amount i między income a property_value

```python
#Przygladam się blizej zmiennym o najwiekszej korelacji

plt.figure(figsize=(8, 6))
sns.scatterplot(
    x='loan_amount', 
    y='property_value', 
    hue='Status',
    data=data, 
    alpha=0.5
)
plt.title('Loan Amount vs. property_value')
plt.xlabel('Loan Amount')
plt.ylabel('property_value')
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(
    x='Interest_rate_spread', 
    y='rate_of_interest', 
    hue='Status',
    data=data, 
    alpha=0.5
)
plt.title('Interest_rate_spread vs. rate_of_interest')
plt.xlabel('Interest_rate_spread')
plt.ylabel('rate_of_interest')
plt.show()
```
